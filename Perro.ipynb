```python
import cv2
import numpy as np
from matplotlib import pyplot as plt 
```

#### Ejercicio

```python

image = cv2.imread('./img/perro.jpg', cv2.IMREAD_GRAYSCALE)


# Filtro de Detección de Bordes (Sobel) 3 1
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=21)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=11)
gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Imagen Original')
plt.subplot(1, 2, 2)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Magnitud del Gradiente')
plt.show()



# Filtro de Umbralización (Thresholding) 135 244
_, thresholded = cv2.threshold(image, 80, 150, cv2.THRESH_BINARY)
plt.subplot(1, 2, 1)
plt.imshow(image, cmap = 'gray')
plt.title('Imagen Original')
plt.subplot(1, 2, 2)
plt.imshow(thresholded, cmap='gray')
plt.title('Imagen Umbralizada')
plt.show()

cv2.imwrite('./img/perrohuesudo.jpg', thresholded)
```

```python
# Leer la imagen
img = cv2.imread('./img/perrohuesudo.jpg', cv2.IMREAD_GRAYSCALE)  # Leer en escala de grises directamente

# Calcular el histograma
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

# Visualizar la imagen y su histograma
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Mostrar la imagen en escala de grises
ax[0].imshow(img, cmap='gray')
ax[0].set_title("Imagen en Grises")
ax[0].axis('off')

# Mostrar el histograma
ax[1].plot(hist, color='black')
ax[1].set_title("Histograma")
ax[1].set_xlabel("Intensidad de Pixel")
ax[1].set_ylabel("Cantidad de Píxeles")
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
img = cv2.imread('./img/perrohuesudo.jpg', cv2.IMREAD_GRAYSCALE)

# Definir el umbral
umbral = 110  # Puedes cambiar este valor según tus necesidades

# Binarizar la imagen
_, img_binarizada = cv2.threshold(img, umbral, 255, cv2.THRESH_BINARY)

# Visualizar la imagen original y la imagen binarizada
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Mostrar la imagen en escala de grises
ax[0].imshow(img, cmap='gray')
ax[0].set_title("Imagen Original")
ax[0].axis('off')

# Mostrar la imagen binarizada
ax[1].imshow(img_binarizada, cmap='gray')
ax[1].set_title(f"Perro huesudo")
ax[1].axis('off')


plt.tight_layout()
plt.show()
```
